```python
import random

list = ["Rock", "Paper", "Scissor"]
uscore = 0
cscore = 0

while True:
    print('''Choose any one
1. Play the Game
2. End''')
    w = int(input("Choose the number from above: "))

    if w == 2:
      print("Game Closed....")
      break

    elif w == 1:
        uscore = 0
        cscore = 0

        for y in range(5):
            user = int(input("Enter number from following choices\n1. Rock\n2. Paper\n3. Scissor: "))

            # User side
            if user == 1:
                uchoice = "Rock"
            elif user == 2:
                uchoice = "Paper"
            elif user == 3:
                uchoice = "Scissor"
            else:
                print("Invalid choice! Try again.")
                continue

            # Computer side
            cchoice = random.choice(list)
            print("You Chose " + uchoice)
            print("Computer Chose " + cchoice)

            # Conditions
            if uchoice == cchoice:
                print("The match is drawn")
                uscore += 1
                cscore += 1

            elif (uchoice == "Rock" and cchoice == "Scissor") or (uchoice == "Paper" and cchoice == "Rock") or (uchoice == "Scissor" and cchoice == "Paper"):
                print("You have got the point")
                uscore += 1

            elif (cchoice == "Rock" and uchoice == "Scissor") or (cchoice == "Paper" and uchoice == "Rock") or (cchoice == "Scissor" and uchoice == "Paper"):
                print("Computer got the point")
                cscore += 1

        if uscore > cscore:
            print("User has won the Game")
        elif cscore > uscore:
            print("Computer has won the Game")
        else:
            print("The Match has been drawn")
        again = input("Do you want to play again? (yes/no): ").strip().lower()
        if again != 'yes':
            print("Game Closed....")
            break
    else:
        print("Invalid choice! Try again.")
```
